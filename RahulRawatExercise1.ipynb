```python
emp = ['Parker','John','Smith','William']
id = [102,107,109,114]
age=[18,19,21,24]
emp_series = pd.Series(emp)
id_series=pd.Series(id)
age_series=pd.Series(age)
frame = {'Emp':emp_series,'ID':id_series,'Age':age_series}
result = pd.DataFrame(frame)
print(result)
```
